```python
#import educational data set
import numpy as np
import pandas as pd
from pandas import read_csv as read
from sklearn import datasets, linear_model, preprocessing

path = "train_data.csv"
data = read(path, sep = ',')

data.head()
```

```python
data.shape
```

```python
# select all categorical data

categorical = data.select_dtypes(include=[object])
categorical.head()
```

```python
# encode labels with value between 0 and n_classes-1.
le = preprocessing.LabelEncoder()

# use df.apply() to apply le.fit_transform to all columns
categorical_2 = categorical.apply(le.fit_transform)

categorical_2.head()
```

```python
#transformation categorical features to numeric using one-hot-encoding with automatic labeling
from category_encoders import OneHotEncoder

OHE = OneHotEncoder(handle_unknown='ignore', use_cat_names=True)

categorical_transf = OHE.fit_transform(categorical)

categorical_transf.head()
```

```python
#shape verification 
categorical_transf.shape
```

```python
#concatination of dataframes: all features + OHE fitted
#it is important to set axis of concatenation! 1 - Y, 0 - X
data_fin = pd.concat([data.iloc[:, 0:144], categorical_transf], axis=1, sort=False)

data_fin.head()
```

```python
#get values: x - features, y - classes
X = data_fin
Y = data.iloc[:, 145:146]
```

```python
from sklearn.cross_validation import train_test_split as train
x_train, x_test, y_train, y_test = train(X, Y, test_size=0.3)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# fit a model
lm = linear_model.LinearRegression()

model = lm.fit(x_train, y_train)
predictions = lm.predict(x_test)
```

```python
predictions[0:5]
```

```python
## The line / model
from matplotlib import pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

```python
print('Score:', model.score(x_test, y_test))
```
